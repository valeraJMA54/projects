```python
import pandas as pd
import numpy as np
import seaborn as sns
import os, time, json, pickle 
import matplotlib.pyplot as plt
```

```python
youtube=pd.read_csv('D:/datasets/RU_youtube_trending_data.csv') 
df=youtube.copy(deep=True)
df
```

```python
df.describe(include='all')
```

```python
with open("D:/datasets/RU_category_id.json", "r") as f:
    cates = json.load(f)
f.close()
    
id2cate = {}
for cate in cates["items"]:
    ids = int(cate["id"])
    snippet = cate["snippet"]
    if ids not in id2cate:
        id2cate[ids] = snippet["title"]
id2cate
```

```python
df["category"] = df.categoryId.apply(lambda x: id2cate[x] if x in id2cate else "Unknown")
df#—É–¥–∞–ª–∏—Ç—å —Å—Ç–æ–ª–±—Ü—ã
```

```python
df["publishedAt"] = pd.to_datetime(df.publishedAt)
df["year"] = df.publishedAt.dt.year 
df["month"] = df.publishedAt.dt.month 
df["day"] = df.publishedAt.dt.day 
df["datetime"] = df.year.astype(str) + "-" + df.month.astype(str) + "-" + df.day.astype(str)
df.drop(["publishedAt", "year", "month", "day"], axis=1, inplace=True)
df["datetime"] = pd.to_datetime(df.datetime)

df[["datetime"]].dtypes
df.head(10)
```

```python
def period(p1,p2):
    period = df[(df['datetime'] >= p1)&(df['datetime'] <= p2)][['datetime','channelTitle']] 
    return period

period('2021-01-12','2021-02-12').value_counts().head(10)
```
